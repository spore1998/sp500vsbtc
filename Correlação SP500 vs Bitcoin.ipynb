```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
btc = pd.read_csv('btc.csv',index_col='Date',parse_dates=True, decimal='.', thousands=',')
btc = btc['Price']

spy = pd.read_csv('spy.csv',index_col='Date',parse_dates=True, decimal='.', thousands=',')
spy = spy['Price']
btc
```

```python
teste = pd.concat([btc,spy],axis=1).dropna()
teste.columns = ['btc','spy']

teste1 = teste.pct_change(1)
teste1

```

```python
teste1.plot.scatter(x='btc', y='spy', c='blue',s=150,alpha=0.3,figsize=(20,10),label='1 dia')
plt.axvline(x=0, color='r')
plt.axhline(y=0, color='r')
```

```python
teste1.corr()
```

```python
teste30 = teste.pct_change(30)
teste30
```
